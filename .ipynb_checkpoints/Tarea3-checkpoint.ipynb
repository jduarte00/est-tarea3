```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy import stats
from scipy.stats import semicircular
from statsmodels.distributions.empirical_distribution import ECDF
from matplotlib.pyplot import figure

np.random.seed(1)
```

## Ejercicio 1

```python
todas_diff = []
for i in range(0, 1000):
  mat_h = np.random.randn(100,100)
  mat_h_s = (mat_h + mat_h.T) / 2
  eigenvals = np.linalg.eigvals(mat_h_s)
  eigenvals[::-1].sort()
  difs = np.diff(eigenvals) * -1
  todas_diff.append(difs)
  
```

```python
todas_diff = np.array(todas_diff)
todas_diff = todas_diff.flatten()
todas_diff = todas_diff / np.mean(todas_diff)
#todas_diff = (todas_diff - np.mean(todas_diff)) / np.std(todas_diff)
```

```python
rango = np.linspace(np.min(todas_diff), np.max(todas_diff), 200)
wigner = (np.pi *rango /2.0)*np.exp((- np.pi * (rango ** 2))/(4.0))
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
sns.histplot(data = todas_diff, stat = "density", element = "step", fill = False, label="Dist. Emp",common_norm=False)

ax = plt.axes()
ax.set_xlabel('Dif entre valores propios consecutivos')
plt.title("Dist. Emp de diferencias de valores propios para GOE y conjetura de Wigner")
plt.plot(rango, wigner,label=r"Wigner")
plt.legend(loc="upper left", bbox_to_anchor = (1,.975))
```

```python
# Se obtiene analíticamente la integral de 0 a x de la densidad dada por la conjetura de wigner

def wigner_cum(x):
    return np.exp(- np.pi *(x **2)/4) * (np.exp(np.pi * (x**2) /4) -1)
  
```

```python
# sSe puede obtener la distribución de prob empírica para las diferencias de los eigenvalores que han sido simulados
dist_emp = ECDF(todas_diff)
```

```python
# Se imprime la prob de que la diferencia sea igual o menor a uno según la distribución acumulada de la conjetura de wigner
wigner_cum(1)
```

```python
# Se imprime la prob de que la diferencia sea igual o menor a uno según la distribución empírica
dist_emp(1)
```

Como se puede observar arriba, la diferencia no es tan pequeña, de hecho el test de Kolmogorov-Smirnov podría rechazar que la muestra venga de la distribución especificada por la conjetura de wigner, como se demuestra a continuación, donde se tiene un p-valor muy pequeño.

```python
stats.kstest( todas_diff, wigner_cum)
```

Sim embargo, si se toma una muestra aleatoria de 500 valores de la distribución especificada por la conjetura de Wigner y se realiza el test de Kolmogorov-Smirnov, no se rechaza la hipótesis nula, es decir, el test afirma que ambas muestras vienen de la misma distribución




```python
def sacar_aleatorios(pdf, interval, pdfmax = 1, integers = False, max_iterations = 10000):
        for i in range(max_iterations):
            if integers == True:
                rand_x = np.random.randint(interval[0], interval[1])
            else:
                rand_x = (interval[1] - interval[0]) * np.random.random(1) + interval[0] #(b - a) * random_sample() + a

            rand_y = pdfmax * np.random.random(1) 
            calc_y = pdf(rand_x)

            if(rand_y <= calc_y ):
                return rand_x

        raise Exception("Could not find a matching random number within pdf in " + max_iterations + " iterations.")
```
